```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
TY = pd.read_csv('orders for black friday.csv')
TY.head(1)
```

```python
TY.info()
```

```python
LY = pd.read_csv('orders for black friday 2017.csv')
```

```python
sns.violinplot('sales_gbp_esiv',data=TY)
```

```python
TY_99 = TY[TY['sales_gbp_esiv']<91.51]
TY_99 = TY_99[TY_99['sales_gbp_esiv']>0.00]
```

```python
LY_99 = LY[LY['sales_gbp_esiv']<106.48]
LY_99 = LY_99[LY_99['sales_gbp_esiv']>0.00]
```

```python
plt.figure(figsize=(12,10))
sns.violinplot('date_key','sales_gbp_esiv',data=TY_99)
```

```python
plt.figure(figsize=(12,10))
sns.boxplot('date_key','sales_gbp_esiv',data=TY_99,whis=1,notch=True)
```

```python
sns.violinplot('sales_gbp_esiv',data=LY_99)
```

```python
plt.figure(figsize=(12,10))
sns.boxplot('date_key','sales_gbp_esiv',data=LY_99)
```
