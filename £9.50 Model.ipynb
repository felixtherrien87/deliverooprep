```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('holidaysdata.csv',index_col=0)
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
df.head()
```

```python
df['max_year'] = pd.to_datetime(df['max_date'])
df.head(5)
df['year'] = df['max_year'].dt.year
df.drop(['max_date','max_year'],axis=1,inplace=True)
```

```python
df.replace(to_replace=1970,value=0)
```

```python
target = df['booked']
labels = df.drop(['booked','created_date'],axis=1)
df.groupby('booked').count()
```

```python
hot_labels = pd.get_dummies(labels[['channel_mix','year', 'gender', 'age_band']])
```

```python
labels_final = pd.concat([hot_labels,labels[['previously_booked', 'sun_savers', 'opened']]],axis=1)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(labels_final, 
                                                    target, 
                                                    test_size = 0.3, 
                                                    random_state = 30)

print("The training data has {} samples".format(X_train.shape[0]))
print("The test data has {} samples".format(X_test.shape[0]))
```

```python
from sklearn.linear_model import LogisticRegression

#buildmodels
clf_log = LogisticRegression(random_state = 40)
#clf_tree = DecisionTreeClassifier(random_state = 40)
#clf_rand = RandomForestClassifier(random_state = 40)

clf_log.fit(X_train,y_train)
#clf_tree.fit(X_train,y_train)
#clf_rand.fit(X_train,y_train)

train_pred = clf_log.predict(X_train)

log_pred = clf_log.predict(X_test)
#tree_pred = clf_tree.predict(X_test)
#rand_pred = clf_rand.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,train_pred))
```
